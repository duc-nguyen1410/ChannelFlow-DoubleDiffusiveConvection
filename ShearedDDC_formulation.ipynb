# Governing equations of Double-diffusive convection
Dimensional governing equations of double-diffusive convection:
\begin{equation}
\begin{aligned}
\nabla\cdot \boldsymbol{U}^* &= 0,\\
\partial_t \boldsymbol{U}^*+\boldsymbol{U}^*\cdot\nabla\boldsymbol{U}^* &= -\nabla P^* + \nu\nabla^2\boldsymbol{U}^* + \mathbf{g}\left(\beta\Delta S^* - \alpha\Delta T^*\right),\\
\partial_t T^* +\boldsymbol{U}^*\cdot\nabla T^* &= \kappa_T\nabla^2 T^*,\\
\partial_t S^* +\boldsymbol{U}^*\cdot\nabla S^* &= \kappa_S\nabla^2 S^*.
\end{aligned}
\end{equation}
with $\mathbf{g}=g\hat{\mathbf{g}}=g(-\sin{\gamma}\hat{\mathbf{e}}_x-\cos{\gamma}\hat{\mathbf{e}}_y)$. So we can rewrite these equations into:
\begin{align}
\nabla\cdot \boldsymbol{U}^* &= 0,\\
\partial_t \boldsymbol{U}^*+\boldsymbol{U}^*\cdot\nabla\boldsymbol{U}^* &= -\nabla P^* + \nu\nabla^2\boldsymbol{U}^* - g\left(\beta\Delta S^* - \alpha\Delta T^*\right)(\sin{\gamma}\hat{\mathbf{e}}_x+\cos{\gamma}\hat{\mathbf{e}}_y),\\
\partial_t T^* +\boldsymbol{U}^*\cdot\nabla T^* &= \kappa_T\nabla^2 T^*,\\
\partial_t S^* +\boldsymbol{U}^*\cdot\nabla S^* &= \kappa_S\nabla^2 S^*.
\end{align}

\begin{align}
\nabla\cdot \boldsymbol{U}^* &= 0,\\
\partial_t \boldsymbol{U}^*+\boldsymbol{U}^*\cdot\nabla\boldsymbol{U}^* &= -\nabla P^* + \nu\nabla^2\boldsymbol{U}^* + g\left(\alpha\Delta T^* - \beta\Delta S^*\right)(\sin{\gamma}\hat{\mathbf{e}}_x+\cos{\gamma}\hat{\mathbf{e}}_y),\\
\partial_t T^* +\boldsymbol{U}^*\cdot\nabla T^* &= \kappa_T\nabla^2 T^*,\\
\partial_t S^* +\boldsymbol{U}^*\cdot\nabla S^* &= \kappa_S\nabla^2 S^*.
\end{align}

\begin{align}
\left[\frac{\sqrt{g\alpha\Delta_T H}}{H}\right]\nabla\cdot \boldsymbol{U} &= 0,\\
\left[\frac{\sqrt{g\alpha\Delta_T H}}{H/\sqrt{g\alpha\Delta_T H}}\right]\partial_t \boldsymbol{U}+\left[\frac{\sqrt{g\alpha\Delta_T H}^2}{H}\right]\boldsymbol{U}\cdot\nabla\boldsymbol{U} &= -\nabla P^* + \left[\frac{\nu\sqrt{g\alpha\Delta_T H}}{H^2}\right]\nabla^2\boldsymbol{U} + \left[g\alpha\Delta_T\right]\left( T - \left[\frac{\beta\Delta_S}{\alpha\Delta_T}\right] S\right)(\sin{\gamma}\hat{\mathbf{e}}_x+\cos{\gamma}\hat{\mathbf{e}}_y),\\
\left[\frac{\Delta_T}{H/\sqrt{g\alpha\Delta_T H}}\right]\partial_t T +\left[\frac{\Delta_T\sqrt{g\alpha\Delta_T H}}{H}\right]\boldsymbol{U}\cdot\nabla T &= \left[\frac{\Delta_T\kappa_T}{H^2}\right]\nabla^2 T,\\
\left[\frac{\Delta_S}{H/\sqrt{g\alpha\Delta_T H}}\right]\partial_t S +\left[\frac{\Delta_S\sqrt{g\alpha\Delta_T H}}{H}\right]\boldsymbol{U}\cdot\nabla S &= \left[\frac{\Delta_S\kappa_S}{H^2}\right]\nabla^2 S.
\end{align}

\begin{align}
\nabla\cdot \boldsymbol{U} &= 0,\\
\partial_t \boldsymbol{U}+\boldsymbol{U}\cdot\nabla\boldsymbol{U} &= -\nabla P + \left[\frac{\nu}{\sqrt{g\alpha\Delta_T H^3}}\right]\nabla^2\boldsymbol{U} + \left( T - \Lambda S\right)(\sin{\gamma}\hat{\mathbf{e}}_x+\cos{\gamma}\hat{\mathbf{e}}_y),\\
\partial_t T +\boldsymbol{U}\cdot\nabla T &= \left[\frac{\kappa_T}{\sqrt{g\alpha\Delta_T H^3}}\right]\nabla^2 T,\\
\partial_t S +\boldsymbol{U}\cdot\nabla S &= \left[\frac{\kappa_S}{\sqrt{g\alpha\Delta_T H^3}}\right]\nabla^2 S.
\end{align}

\begin{align}
\nabla\cdot \boldsymbol{U} &= 0,\\
\partial_t \boldsymbol{U}+\boldsymbol{U}\cdot\nabla\boldsymbol{U} &= -\nabla P + \left[\sqrt{\frac{\nu\kappa_T}{g\alpha\Delta_T H^3}}\sqrt{\frac{\nu}{\kappa_T}}\right]\nabla^2\boldsymbol{U} + \left( T - \Lambda S\right)(\sin{\gamma}\hat{\mathbf{e}}_x+\cos{\gamma}\hat{\mathbf{e}}_y),\\
\partial_t T +\boldsymbol{U}\cdot\nabla T &= \left[\sqrt{\frac{\nu\kappa_T}{g\alpha\Delta_T H^3}}\sqrt{\frac{\kappa_T}{\nu}}\right]\nabla^2 T,\\
\partial_t S +\boldsymbol{U}\cdot\nabla S &= \left[\sqrt{\frac{\nu\kappa_T}{g\alpha\Delta_T H^3}}\sqrt{\frac{\kappa_T}{\nu}}\frac{\kappa_S}{\kappa_T}\right]\nabla^2 S.
\end{align}

Thus, nondimensional governing equations of sheared double-diffusive convection:
\begin{align}
\nabla\cdot \boldsymbol{U} &= 0,\\
\partial_t \boldsymbol{U}+\boldsymbol{U}\cdot\nabla\boldsymbol{U} &= -\nabla P + \sqrt{\frac{Pr}{Ra}}\nabla^2\boldsymbol{U} + \left(T-\Lambda S\right)(\sin{\gamma}\hat{\mathbf{e}}_x+\cos{\gamma}\hat{\mathbf{e}}_y),\\
\partial_t T +\boldsymbol{U}\cdot\nabla T &= \frac{1}{\sqrt{PrRa}}\nabla^2 T,\\
\partial_t S +\boldsymbol{U}\cdot\nabla S &= \frac{1}{Le\sqrt{PrRa}}\nabla^2 S.
\end{align}
where
\begin{equation}
Ra=\frac{\nu}{\kappa_T}, \quad Ra=\frac{g\alpha\Delta_T H^3}{\nu\kappa_T}, \quad R_\rho=\frac{\alpha\Delta_T}{\beta\Delta_S}=\frac{1}{\Lambda}, \quad Ri=\frac{g\alpha\Delta_T H}{U^2_b}(\Lambda-1), \quad Le=\frac{\kappa_T}{\kappa_S}
\end{equation}

This system was nondimensionalized by height length $H$, background scalar differences $\Delta_T$ and $\Delta_S$, and free-fall velocity $\sqrt{g\alpha\Delta_T H}$

Another form of governing equations based on velocity of boundaries $U_b$:
\begin{align}
\nabla\cdot \boldsymbol{U} &= 0,\\
\partial_t \boldsymbol{U}+\boldsymbol{U}\cdot\nabla\boldsymbol{U} &= -\nabla P + \frac{1}{Re}\nabla^2\boldsymbol{U} + \frac{Ri_b}{\Lambda-1}\left(T-\Lambda S\right)(\sin{\gamma}\hat{\mathbf{e}}_x+\cos{\gamma}\hat{\mathbf{e}}_y),\\
\partial_t T +\boldsymbol{U}\cdot\nabla T &= \frac{1}{RePr}\nabla^2 T,\\
\partial_t S +\boldsymbol{U}\cdot\nabla S &= \frac{1}{LeRePr}\nabla^2 S.
\end{align}

## Case $\gamma=0$: Horizontal channel
With $\gamma=0$, the nondimensional governing equations can be reduced into:
\begin{equation}
\begin{aligned}
\nabla\cdot \boldsymbol{U} &= 0,\\
\partial_t \boldsymbol{U}+\boldsymbol{U}\cdot\nabla\boldsymbol{U} &= -\nabla P + \sqrt{\frac{Pr}{Ra}}\nabla^2\boldsymbol{U} + \left(T-\Lambda S\right)\hat{\mathbf{e}}_y,\\
\partial_t T +\boldsymbol{U}\cdot\nabla T &= \frac{1}{\sqrt{PrRa}}\nabla^2 T,\\
\partial_t S +\boldsymbol{U}\cdot\nabla S &= \frac{1}{Le\sqrt{PrRa}}\nabla^2 S.
\end{aligned}
\end{equation}
This is a typical case, which was used in many previous studies (Yang et al. 2021, 2022, 2024)

## Case $\gamma\neq0$: Inclined channel
The nondimensional governing equations:
\begin{equation}
\begin{aligned}
\nabla\cdot \boldsymbol{U} &= 0,\\
\partial_t \boldsymbol{U}+\boldsymbol{U}\cdot\nabla\boldsymbol{U} &= -\nabla P + \sqrt{\frac{Pr}{Ra}}\nabla^2\boldsymbol{U} + \left(T-\Lambda S\right)(\sin{\gamma}\hat{\mathbf{e}}_x+\cos{\gamma}\hat{\mathbf{e}}_y),\\
\partial_t T +\boldsymbol{U}\cdot\nabla T &= \frac{1}{\sqrt{PrRa}}\nabla^2 T,\\
\partial_t S +\boldsymbol{U}\cdot\nabla S &= \frac{1}{Le\sqrt{PrRa}}\nabla^2 S.
\end{aligned}
\end{equation}

To determine base flow profiles, the perturbations are assumed to be zero
\begin{align}
    U: \quad 0 &= -\Pi_x(y) &+ \sqrt{\frac{Pr}{Ra}} \frac{\partial^2}{\partial y^2} U_{0}(y) &+\left(T_0(y) - \Lambda S_0(y)\right)\sin{\gamma},\\
    V: \quad 0 &= -\Pi_y(y) &&+ \left(T_0(y) - \Lambda S_0(y)\right)\cos{\gamma},\\
    W: \quad 0 &= -\Pi_z(y)  &+ \sqrt{\frac{Pr}{Ra}} \frac{\partial^2}{\partial y^2} W_{0}(y) &,\\
    T: \quad 0 &=  &\frac{1}{\sqrt{PrRa}}\frac{\partial^2}{\partial y^2} T_0(y) &,\\
    S: \quad 0 &=  &\frac{1}{Le\sqrt{PrRa}}\frac{\partial^2}{\partial y^2} S_0(y) &,
\end{align}

1. Laminar temperature profile:
   $$T_0(y) = \frac{T_b - T_a}{b-a} y + \frac{bT_a - aT_b}{b-a}$$
2. Laminar salinity profile:
   $$S_0(y) = \frac{S_b - S_a}{b-a} y + \frac{bS_a - aS_b}{b-a}$$
3. Pressure gradient profiles:
   $$\Pi_x(y)=\sqrt{\frac{Pr}{Ra}} \frac{\partial^2}{\partial y^2} U_{0}(y) +\left(T_0(y) - \Lambda S_0(y)\right)\sin{\gamma},$$
   $$\Pi_y(y)=\left(T_0(y) - \Lambda S_0(y)\right)\cos{\gamma},$$
   $$\Pi_z(y)=\sqrt{\frac{Pr}{Ra}} \frac{\partial^2}{\partial y^2} W_{0}(y).$$
   - Hydrostatic case (requires stationary walls): $U_{0}(y)=W_{0}(y)=0$
      - $\gamma=0$: 
         $$\Pi_x(y)=0,$$
         $$\Pi_y(y)=\left(T_0(y) - \Lambda S_0(y)\right),$$
         $$\Pi_z(y)=0.$$
      - $\gamma\neq 0$: 
         $$\Pi_x(y)=\left(T_0(y) - \Lambda S_0(y)\right)\sin{\gamma},$$
         $$\Pi_y(y)=\left(T_0(y) - \Lambda S_0(y)\right)\cos{\gamma},$$
         $$\Pi_z(y)=0.$$
      - general selection:
         $$\Pi_x(y)=0,$$
         $$\Pi_y(y)=\left(T_0(y) - \Lambda S_0(y)\right)\cos{\gamma},$$
         $$\Pi_z(y)=0.$$
4. Laminar velocity profiles:
   Based on pressure gradient, temperature, and salinity profiles, laminar velocity profiles can be obtained from
   $$\sqrt{\frac{Pr}{Ra}} \frac{\partial^2}{\partial y^2} U_{0}(y) = -\left(T_0(y) - \Lambda S_0(y)\right)\sin{\gamma},$$
   $$W_0(y) = \frac{W_b - W_a}{b-a} y + \frac{bW_a - aW_b}{b-a},$$


* Determining x-velocity base profile: 
    $$p_1 \frac{\partial^2}{\partial y^2} U_{0}(y) = -p_2\left(p_3 T_0(y) - p_4 S_0(y)\right)\sin{\gamma},$$
    $$-\frac{p_1}{p_2 \sin{\gamma}} \frac{\partial^2}{\partial y^2} U_{0}(y) = p_3 (p_{T,1} y + p_{T,0}) - p_4 (p_{S,1} y + p_{S,0}),$$
    $$-\frac{p_1}{p_2 \sin{\gamma}} \frac{\partial^2}{\partial y^2} U_{0}(y) = (p_3 p_{T,1} - p_4 p_{S,1}) y + (p_3 p_{T,0} - p_4 p_{S,0}),$$
    $$-\frac{p_1}{p_2 \sin{\gamma}} U_{0}(y) = \frac{1}{6}(p_3 p_{T,1} - p_4 p_{S,1}) y^3 + \frac{1}{2}(p_3 p_{T,0} - p_4 p_{S,0}) y^2 + C_1 y + C_2,$$
    * $y=a$: 
        $$-\frac{p_1}{p_2 \sin{\gamma}} U_a = \frac{1}{6}(p_3 p_{T,1} - p_4 p_{S,1}) a^3 + \frac{1}{2}(p_3 p_{T,0} - p_4 p_{S,0}) a^2 + C_1 a + C_2,$$
    * $y=b$: 
        $$-\frac{p_1}{p_2 \sin{\gamma}} U_b = \frac{1}{6}(p_3 p_{T,1} - p_4 p_{S,1}) b^3 + \frac{1}{2}(p_3 p_{T,0} - p_4 p_{S,0}) b^2 + C_1 b + C_2,$$
    thus,
    $$-\frac{p_1}{p_2 \sin{\gamma}} (U_b-U_a) = \frac{1}{6}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3-a^3) + \frac{1}{2}(p_3 p_{T,0} - p_4 p_{S,0}) (b^2-a^2) + C_1 (b-a),$$
    $$C_1 = -\frac{p_1}{p_2 \sin{\gamma} (b-a)} (U_b-U_a) - \frac{1}{2(b-a)}\left[\frac{1}{3}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3-a^3) + (p_3 p_{T,0} - p_4 p_{S,0}) (b^2-a^2)\right],$$
    and
    $$-\frac{p_1}{p_2 \sin{\gamma}} (U_b+U_a) = \frac{1}{6}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3+a^3) + \frac{1}{2}(p_3 p_{T,0} - p_4 p_{S,0}) (b^2+a^2) + C_1 (b+a) + 2C_2,$$
    $$C_2 = -\frac{p_1}{2 p_2 \sin{\gamma}} (U_b+U_a) - \frac{1}{12}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3+a^3) - \frac{1}{4}(p_3 p_{T,0} - p_4 p_{S,0}) (b^2+a^2) \\- \frac{b+a}{2} \left(-\frac{p_1}{p_2 \sin{\gamma} (b-a)} (U_b-U_a) - \frac{1}{2(b-a)}\left[\frac{1}{3}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3-a^3) + (p_3 p_{T,0} - p_4 p_{S,0}) (b^2-a^2)\right]\right),$$

$$C_2 = -\frac{p_1}{2 p_2 \sin{\gamma}} \left[(U_b+U_a)-(U_b-U_a)\frac{b+a}{b-a}\right] - \frac{1}{12}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3+a^3) - \frac{1}{4}(p_3 p_{T,0} - p_4 p_{S,0}) (b^2+a^2) \\+\frac{b+a}{4(b-a)}\left[\frac{1}{3}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3-a^3) + (p_3 p_{T,0} - p_4 p_{S,0}) (b^2-a^2)\right],$$

$$C_2 = -\frac{p_1}{2 p_2 \sin{\gamma}} \left[(U_b+U_a)-(U_b-U_a)\frac{b+a}{b-a}\right] \\
- \frac{1}{12}(p_3 p_{T,1} - p_4 p_{S,1}) \left[(b^3+a^3) - (b^3-a^3)\frac{b+a}{b-a}\right]\\
-\frac{1}{4}(p_3 p_{T,0} - p_4 p_{S,0}) \left[(b^2+a^2) - (b^2-a^2)\frac{b+a}{b-a}\right],$$

Final laminar velocity profile:
$$U_{0}(y) = p_{u3} y^3 + p_{u2} y^2 + p_{u1} y + p_{u0},$$
with
$$p_{u3} = -\frac{p_2 \sin{\gamma}}{6 p_1}(p_3 p_{T,1} - p_4 p_{S,1})$$
$$p_{u2} = -\frac{p_2 \sin{\gamma}}{2 p_1}(p_3 p_{T,0} - p_4 p_{S,0})$$
$$p_{u1} = -\frac{p_2 \sin{\gamma}}{p_1}C_1\\
=-\frac{p_2 \sin{\gamma}}{p_1} \left(-\frac{p_1}{p_2 \sin{\gamma} (b-a)} (U_b-U_a) - \frac{1}{2(b-a)}\left[\frac{1}{3}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3-a^3) + (p_3 p_{T,0} - p_4 p_{S,0}) (b^2-a^2)\right]\right)\\
=\frac{U_b-U_a}{b-a} + \frac{p_2 \sin{\gamma}}{2p_1(b-a)}\left[\frac{1}{3}(p_3 p_{T,1} - p_4 p_{S,1}) (b^3-a^3) + (p_3 p_{T,0} - p_4 p_{S,0}) (b^2-a^2)\right],$$
$$p_{u0} = -\frac{p_2 \sin{\gamma}}{p_1}C_2\\
= \frac{1}{2} \left[(U_b+U_a)-(U_b-U_a)\frac{b+a}{b-a}\right] \\
+\frac{p_2 \sin{\gamma}}{12p_1} (p_3 p_{T,1} - p_4 p_{S,1}) \left[(b^3+a^3) - (b^3-a^3)\frac{b+a}{b-a}\right]\\
+\frac{p_2 \sin{\gamma}}{4p_1} (p_3 p_{T,0} - p_4 p_{S,0}) \left[(b^2+a^2) - (b^2-a^2)\frac{b+a}{b-a}\right],$$

````python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['mathtext.fontset'] = 'stix'  # Use STIX fonts for math symbols
plt.rcParams['text.usetex'] = True # enable LaTeX rendering
def ypoints(a, b, Ny):
    ypts = np.zeros(Ny)
    c = 0.5 * (b + a)
    r = 0.5 * (b - a)
    piN = np.pi / (Ny - 1)
    for i in range(Ny):
        ypts[i] = c + r * np.cos(piN * i)
    return ypts

# Define dimensionless parameters
y, Pr, Re, Ra, Le, Ri, Rrho = sp.symbols('y Pr Re Ra Le Ri Rrho')

# Define controlling parameters of problem
P1 = (Pr/Ra)**(0.5)
P2 = 1.0
P3 = 1.0
P4 = Rrho
P5 = (Pr*Ra)**(-0.5)
P6 = (Pr*Ra)**(-0.5) / Le

# Define the parameters of compudational domain
a, b = -0.5, 0.5 # ymin/ymax
Ua, Ub = 0, 0 # boundary condition of x-velocity
Wa, Wb = 0, 0 # boundary condition of z-velocity
Ta, Tb = 0.5, -0.5 # boundary condition of temperature
Sa, Sb = 0.5, -0.5 # boundary condition of salinity
gammax = 45.0 # inclined angle

# subparameters
h = b-a
c = h/2.0
d = (b+a)/2.0
minor2 = b*b-a*a
minor3 = b*b*b-a*a*a
plus2 = b*b+a*a
plus3 = b*b*b+a*a*a
sgammax = np.sin(gammax/180.0*np.pi)
cgammax = np.cos(gammax/180.0*np.pi)

############## laminar base profile of temperature #################
### $$T_0(y) = \frac{T_b - T_a}{b-a} y + \frac{bT_a - aT_b}{b-a} = pt1*y + pt0$$
pt1 = (Tb-Ta)/h
pt0 = (b*Ta-a*Tb)/h
Tbase = pt1 * y + pt0
print(f'Laminar base profile of temperature: T_base(y) = {Tbase}')
####################################################################

############## laminar base profile of salinity #################
### $$S_0(y) = \frac{S_b - S_a}{b-a} y + \frac{bS_a - aS_b}{b-a} = pt1*y + pt0$$
ps1 = (Sb-Sa)/h
ps0 = (b*Sa-a*Sb)/h
Sbase = ps1 * y + ps0
print(f'Laminar base profile of salinity: S_base(y) = {Sbase}')
####################################################################

############## laminar base profile of pressure gradient #################
### $$S_0(y) = \frac{S_b - S_a}{b-a} y + \frac{bS_a - aS_b}{b-a} = pt1*y + pt0$$
dPdybase = P2*(P3*Tbase-P4*Sbase)*cgammax
print(f'Laminar base profile of pressure gradient: dPdx_base(y) = 0')
print(f'                                           dPdy_base(y) = {dPdybase}')
print(f'                                           dPdz_base(y) = 0')
####################################################################

############## laminar base profile of x-velocity #################
### 
pu3 = -1.0*P2*sgammax/(6.0*P1) * (P3*pt1-P4*ps1)
pu2 = -1.0*P2*sgammax/(2.0*P1) * (P3*pt0-P4*ps0)
pu1 = (Ub-Ua)/h \
    + P2*sgammax/(2.0*P1*h)*(1.0/3.0*(P3*pt1-P4*ps1)*minor3 + (P3*pt0-P4*ps0)*minor2)
pu0 = 0.5*((Ub+Ua)-(Ub-Ua)*(b+a)/h)\
            + P2*sgammax/(12.0*P1)*(P3*pt1-P4*ps1)*(plus3-minor3*(b+a)/h)\
            + P2*sgammax/(4.0*P1)*(P3*pt0-P4*ps0)*(plus2-minor2*(b+a)/h)
Ubase = pu3 * y**3 + pu2 * y**2 + pu1 * y + pu0
print(f'Laminar base profile of x-velocity: U_base(y) = {Ubase}')
####################################################################


# Define the values for y and parameters
y_values = ypoints(a,b,100)  # Example values for y
params = {Pr: 7, Re: 400, Ra: 1e4, Le: 100, Ri: 1.0, Rrho: 2}
Tbase_values = [Tbase.subs({y: val, **params}) for val in y_values]
Sbase_values = [Sbase.subs({y: val, **params}) for val in y_values]
Ubase_values = [Ubase.subs({y: val, **params}) for val in y_values]

plt.figure(figsize=(2, 3))
plt.plot(Tbase_values,y_values,color='tab:green',label=r'$T_b$')
plt.plot(Sbase_values,y_values,color='tab:red', linestyle = 'dashed',label=r'$S_b$')
plt.plot(Ubase_values,y_values,color='k',label=r'$U_b$')
# plt.xlabel(fontsize=14)
plt.ylabel(r'$y$', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks([a,0,b],fontsize=14)
# plt.xlim([4000, 10200])
plt.ylim([a, b])
plt.legend()
# plt.grid(True)
# plt.savefig('base_profiles.png', dpi=300, bbox_inches='tight')
plt.show()
````

## Chebyshev coefficients of base profiles
In the Channelflow, a series of Chebyshev polynomials is used to represent laminar base profiles in the wall-normal direction.  The first four Chebyshev polynomials of the first kind for the reference interval $[-1,1]$ are 
\begin{align}
        \mathcal{C}_0(\xi) &= 1, \\
        \mathcal{C}_1(\xi) &= \xi, \\
        \mathcal{C}_2(\xi) &= 2\xi^2 - 1, \\
        \mathcal{C}_3(\xi) &= 4\xi^3 - 3\xi,
\end{align}
with the normalization $\mathcal{C}_j(\xi=1)=1$. The polynomial form of laminar base solutions is expressed as:
\begin{equation}
    f(y) = p_3 y^3 + p_2 y^2 + p_1 y + p_0.
\end{equation}
For the general case of walls located at $y_{min}=a$ and $y_{max}=b$, we need a stretching transformation $y=c\xi+d$ with $c=(b-a)/2$ and $d=(b+a)/2$.  A couple of common vertical intervals are the unit interval $y\in[0,1]$ for the double-periodic DDC problem \citep{radko2016thermohaline} and symmetric intervals $y\in[-0.5,0.5]$ for inclined layer convection \citep{reetz2019turbulent}. Thus, the coefficients $q_i$ of general Chebyshev expansion must also satisfy $f(y)=\sum_{i=0}^{3} q_i C_i(\xi)$. By taking homogeneous coefficients, laminar base solutions have the following coefficients: 

\begin{align}
    q_0 &= \left(\frac{3}{2}c^2d+d^3\right) p_3+\left(\frac{1}{2}c^2+d^2\right) p_2+ d p_1 + p_0, \\
    q_1 &= \left(\frac{3}{4}c^3+3cd^2\right)p_3 + 2cd p_2+ c p_1, \\
    q_2 &= \frac{3}{2}c^2 d p_3+\frac{1}{2}c^2 p_2, \\
    q_3 &= \frac{c^3}{4} p_3.
\end{align}

Thus, for example, the streamwise velocity profile $U_0(y)$ has the Chebyshev coefficients
\begin{equation}
    q_0 = \frac{d-a}{b-a}(U_b-U_a)+U_a, \quad
    q_1 = \frac{c}{b-a}(U_b-U_a).
\end{equation}
The Chebyshev coefficients of profiles of spanwise velocity $W_0(y)$, temperature $T_0(y)$, salinity $S_0(y)$ have the same forms with corresponding boundary conditions. The pressure gradient's base profiles $\Pi_y(y)$ have the Chebyshev coefficients linearizing those of temperature and salinity
\begin{equation}
     q_0 = p_2\left(p_3 q_{0,T_0} - p_4 q_{0,S_0} \right), \quad
    q_1 = p_2\left(p_3 q_{1,T_0} - p_4 q_{1,S_0} \right).
\end{equation}

### Boundary conditions
In standard Channelflow, no-slip (Dirichlet) boundary condition was used
\begin{align}
p_1 \boldsymbol{u}''-\lambda \boldsymbol{u} -\nabla q = - \boldsymbol{R},\\
\nabla \cdot \boldsymbol{u} = 0,\\
\boldsymbol{u}(\pm 1) = 0.
\end{align}

The aim of this section is to change noslip (Dirichlet) B.C. to free-slip (Neumann) B.C. 
\begin{align}
p_1 \boldsymbol{u}''-\lambda \boldsymbol{u} -\nabla q = - \boldsymbol{R},\\
\nabla \cdot \boldsymbol{u} = 0,\\
\boldsymbol{u}(\pm 1) = 0, \boldsymbol{u}'(\pm 1) = 0.
\end{align}

# Nondimensional parameters
Structure morphology affects the vertical transport property of system, its sensitivity can be determined by three nondimensional parameters, including Nusselt numbers (for temperature and salt concentration) and vertical Reynolds number: 

$$ Nu_\theta=\left|\frac{\langle w^*\theta^* \rangle_h - \kappa_\theta \partial_z\langle \theta^* \rangle_h}{\kappa_\theta \Delta_\theta H^{-1}}\right| $$
$$ Nu_s \text{(or Sh)}=\left|\frac{\langle w^*s^* \rangle_h - \kappa_s \partial_z\langle s^* \rangle_h}{\kappa_s \Delta_s H^{-1}}\right| $$
$$ Re_z =\frac{w^{rms}H}{\nu} $$

Note that these formulation are nondimensional forms based dimensional quantities of dimensional NSEs.

## How to get these equations?
First, we consider the total temperature equations
$$ \partial_t T^* +\boldsymbol{U}^*\cdot\nabla T^* = \kappa_T\nabla^2 T^* $$
Sine $(\nabla\cdot\boldsymbol{U}^*) T^*+\boldsymbol{U}^*\cdot (\nabla T^*)=\nabla \cdot (\boldsymbol{U}^*T^*)$ by product rule with $(\nabla\cdot\boldsymbol{U}^*) T^*=0$ by continuity equation for incompressible flows, leading to
$$ \partial_t T^* + \nabla \cdot (\boldsymbol{U}^*T^*) = \kappa_T\nabla^2 T^* $$ 
Now we take horizontal average this equations, thus
$$ \partial_t <^*T>_h +<\nabla \cdot (\boldsymbol{U}^*T^*)>_h = \kappa_T\nabla^2 <T^*>_h $$
$$ 0 +\partial_z(<W^* T^*>_h) = \kappa_T \partial_{zz} <T^*>_h $$
$$ \partial_z\left(\langle W^* T^*\rangle_h  - \kappa_T \partial_{z} \langle T^*\rangle_h \right)= 0 $$
by taking integration, then
$$ \langle W^* T^*\rangle_h  - \kappa_T \partial_{z} \langle T^*\rangle_h = C $$
Because this formulas is constant so we can use this as reference parameter, named Nusselt number 

To determine $C$, we can apply the boundary condition at $y=a$ :
$$ \langle W^* T^*\rangle_h|_{y=a}  - \kappa_T \partial_{z} \langle T^*\rangle_h|_{y=a} = - \kappa_T \partial_{z} \langle T^*\rangle_h|_{y=a} = C $$
Thus
$$ \langle W^* T^*\rangle_h  - \kappa_T \partial_{z} \langle T^*\rangle_h = - \kappa_T \partial_{z} \langle T^*\rangle_h|_{y=a} $$


Final, the Nusselt number for temperature is nondimensional parameter by normalizing:
$$ Nu_\theta = \frac{\langle W^* T^*\rangle_h  - \kappa_T \partial_{z} \langle T^*\rangle_h}{\kappa_T \Delta_T H^{-1}} = - \frac{\kappa_T \partial_{z} \langle T^*\rangle_h|_{y=a}}{\kappa_T \Delta_T H^{-1}} $$
and then nondimensional Nusselt number is
$$ Nu_\theta = - \partial_{z} \langle T\rangle_h|_{y=a} $$

Typically, people often use absolute values of the Nussel number:
$$ Nu_\theta = |- \partial_{z} \langle T\rangle_h|_{y=a}| $$

Do similirity for salt concentration, we have
$$ Nu_s \text{ or } Sh = |- \partial_{z} \langle S\rangle_h|_{y=a}| $$
